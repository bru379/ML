# Welcome to Colab!

```
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np
```

##load the data

```
df=pd.read_csv('canada_per_capita_income.csv')
```

```
df.head()
```

## visualize the data

```
%matplotlib inline

plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='*')
plt.xlabel('YEAR')
plt.ylabel('PER CAPITA INCOME')
plt.title('PER CAPITA INCOME IN CANADA')
plt.show()
```

## model training



```
reg=linear_model.LinearRegression()
```

### fit the data

```
reg.fit(df[['year']],df['per capita income (US$)'])
#use double bracket to x features(maintain grid/tabular view)

```

## evaluation

```
# Predict the income for 2020
prediction_2020 = reg.predict(pd.DataFrame({'year': [2020]}))

print(f"Predicted Per Capita Income for 2020: ${prediction_2020[0]:,.2f}")
```
